# Exercício 4

4) Uma universidade tem 10 mil alunos dos quais 4 mil são considerados esportistas. Temos, ainda, que 500 alunos são do curso de biologia diurno, 700 da biologia noturno, 100 são esportistas e da biologia diurno e 200 são esportistas e da biologia noturno. Um aluno é escolhido, ao acaso, e pergunta-se a probabilidade de:

a) Ser esportista? 

R: Se U (Conjunto de todos os eventos) possui 10 mil elementos e E possui 4 mil, teremos:

P(E)=4000/10000 = 4/10 = 40% de chance

```
p=4/10
p
```

b) Ser esportista e aluno da biologia noturno?

R: Seja N o conjunto dos alunos da biologia noturno e esse conjunto possui 700 alunos. Sabemos que B∩E possui 200 alunos, portanto:

P(N∩E) = 200/10000 = 2% de chance

c) Não ser da biologia?

R: Seja B o conjunto de todos os alunos da biologia (diurno e noturno), teremos 1200 alunos no total. Para o aluno escolhido não seja da biologia, teremos que analisar o B complementar, ou seja, U-B:

P(U-B)=10000-1200/10000 = 88% de chance

```
p=(10000-1200)/10000
p
```
