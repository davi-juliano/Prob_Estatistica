```
import pandas as pd
import numpy as np
import statistics #Importações dos pacotes no python
import matplotlib.pyplot as plt
```

```
def Average(lst): #Função de Calcular a Média de uma Lista
    return sum(lst) / len(lst) 

def Ordena_lista(lst): #Função de Ordenação de Listas
  return lst.sort()

def Variancia_populacao(lst):  #Função de Calcular Variância de uma População
  return statistics.pvariance(lst)

def Variancia_amostra(lst):  #Função de Calcular Variância de uma Amostra
  return statistics.variance(lst)

def Desvpad_populacao(lst):  #Função de Calcular Desvio Padrão de uma População
  return statistics.pstdev(lst)

def Desvpad_amostra(lst):  #Função de Calcular Desvio Padrão de uma Amostra
  return statistics.stdev(lst)

def Moda(lst):  #Função de Calcular Moda de uma Lista
  return statistics.mode(lst)

def Mediana(lst):  #Função de Calcular a Mediana de uma Lista
  return statistics.median(lst)

def Tira_cov(a, b):   #Função de Calcular a Covariância entre duas listas

    if len(a) != len(b):
        return

    a_mean = np.mean(a)
    b_mean = np.mean(b)

    sum = 0

    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))

    return sum/(len(a)-1)

def Imc(a, b):
  return a /(b**2)

def descricao_dataframe(data):  #Função de Descrição de um DataFrame do Pandas
  return data.describe()
```

```
import pandas as pd
data = pd.read_excel("IMC_prob.xlsx", index_col=0)

data.describe()

Altura = data['Altura'].values.tolist()

Peso = data['Peso'].values.tolist()

imc=[]

imc_abaixo=[]
imc_normal=[]
imc_acima=[]
imc_obesidade1=[]
imc_obesidade2=[]

for i in range(56):
  imc.append(Imc(Peso[i], Altura[i]))

for i in range(56):
  if (imc[i]<=18.5):
    imc_abaixo.append(imc[i])
  
  if (imc[i]<=25 and imc[i] > 18.5):
    imc_normal.append(imc[i])

  if (imc[i]<= 30 and imc[i] > 25):
    imc_acima.append(imc[i])

  if (imc[i]<= 35 and imc[i] > 30):
    imc_obesidade1.append(imc[i])

  if (imc[i]<= 40 and imc[i] > 35):
    imc_obesidade2.append(imc[i])
        
```

```
Ordena_lista(imc)
numpessoas=[]
for i in range(len(media_peso)):
  numpessoas.append(i)

plt.bar(numpessoas, imc, color='blue')
plt.ylabel("IMC dos Entrevistados")
plt.savefig('IMC.jpg')
plt.show()
```
