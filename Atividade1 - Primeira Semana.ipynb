# Primeiro exercício

```
import statistics as stcs #Importações dos pacotes no python
import matplotlib.pyplot as plt
%matplotlib inline
```

Primeiramente, daremos nome aos bois, ou seja, colocaremos em uma lista cada uma das variáveis:


```
sambaqui=[30.61, 42.88, 27.94, 28.89, 36.22, 41.45, 32.21, 28.86, 42.59, 24.25, 22.56, 15.25, 25.63, 22.92, 33.29]
#Lista de Sambaqui

manguezal=[25.34, 9.49, 19.17, 25.67, 16.92, 21.60, 17.64, 12.91, 20.01, 33.97, 14.05, 19.81, 11.13, 14.88, 16.22]
#Lista do Manguezal
```

Agora, faremos a primeira operação, a média:

```
media_sambaqui = stcs.mean(sambaqui) #Cálculo da média de Sambaqui
print('A média de Sambaqui é %f' %media_sambaqui)

media_manguezal = stcs.mean(manguezal) #Cálculo da média de Manguezal
print('A média de Manguezal é %f' %media_manguezal)
```

Agora iremos fazer a segunda operação, a mediana dos valores:

```
mediana_sambaqui = stcs.median(sambaqui) #Cálculo da mediana de Sambaqui
print('A mediana de Sambaqui é %f' %mediana_sambaqui)

mediana_manguezal = stcs.median(manguezal) #Cálculo da mediana de Manguezal
print('A mediana de Manguezal é %f' %mediana_manguezal)
```

O segundo exercício pede o desvio padrão dos dois locais, portanto:

```
despadrao_sambaqui = stcs.stdev(sambaqui) #Cálculo do desvio padrão de Sambaqui
print('O desvio padrão de Sambaqui é %f' %despadrao_sambaqui)

despadrao_manguezal = stcs.stdev(manguezal) #Cálculo do desvio padrão de Manguezal
print('O desvio padrão de Manguezal é %f' %despadrao_manguezal)
```

# Terceiro Exercício

Para o terceiro exercício, adicionaremos um data set diferente, pois iremos construir um histograma e analisar a sua forma.

Existem diversos tipos de histograma e diversas classificações pra histogramas. 

O primeiro é o **Simétrico**, com a frequência mais alta no centro.

O segundo é o **Assimétrico**, com a frequência mais alta na esquerda ou direita.

O terceiro é o **Despenhadeiro**, com a frequência mais alta em uma das extremidades.

O quarto é o **Dois Picos**, com frequências com picos nas duas extremidades, podendo não ser necessariamente equivalentes, mas devem ser picos.

O quinto é o **Achatado**, onde o gráfico não possui nenhum ponto de destaque, sendo todas as frequências muito equivalentes.

O sexto é o **Pico Isolado**, onde uma das barras possui um destaque além do comum.

```
x =['(8,11]','(11,14]','(14,17]','(17,20]','(20,23]','(23,26]','(26,29]','(29,32]','(32,35]','(35,38]']
y =[3,6,5,7,4,4,2,2,1,1]

plt.bar(x, y, label = 'Barras', color = 'r')
plt.xticks(rotation='45')
plt.legend()
plt.show()
```
