# Exercício 2 - 

2) Dada a variável aleatória X, uniforme em (5, 10), calcule as seguintes probabilidades: 

a) P(X < 7)

P(X < 7)= P(7 > X > 5) = 2/5 = 0.4

2b) P(8 < X < 9)

P(8 < X < 9) = 1/5 = 0.2

2c) P(X > 8,5)

P(X > 8.5) = (10-8.5)/5 = 1.5/5 = 0.3

2d) P(|X - 7,5| > 2) 

P(|X - 7,5| > 2) = P(X < 5.5 ou X > 9.5) = 0.5/5 + 0.5/5 = 0.1 + 0.1 = 0.2

3) Suponha que um mecanismo eletrônico tenha um tempo de vida X (em 1.000 horas) que possa ser considerado uma v.a. contínua com f.d.p. f (x)= e^(-x) , x > 0. O fabricante garante total devolução se X ≤ 0,9. Qual a probabilidade de o item ser devolvido? 

Para calcular, teremos:

integral f(x) = -e^(-x) , aplicando o TFC, teremos:

P(X <= 900) = -e^(-900) + e^(-0) = 1 - e^(-900)
