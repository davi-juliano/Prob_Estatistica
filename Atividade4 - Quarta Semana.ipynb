a) dez ou mais chamadas;

Probabilidade de Poisson é dada por:

P(X=x) = (exp(-lambda)*lambda^(x))/x!, sendo lambda a ocorrencia e o x a variável

Seja X = 10 até X = n, P(X) = somatório(x=10 até x=n)(exp(-8)*8^(x))/x!, sendo n > 10, n em inteiros

b) menos que nove chamadas;

Seja X = 0 até X = 9, P(X) = somatório(x=0 até x=9)(exp(-8)*8^(x))/x!

c) entre sete (inclusive) e nove (exclusive) chamadas.

Seja X = 7 até X = 8, P(X) = somatório(x=7 até x=8)(exp(-8)*8^(x))/x!
