```
x=[163, 150, 178, 180, 171, 180, 162, 180, 177, 165, 168, 159, 171, 166, 180, 167, 168, 167, 160, 186, 185, 167, 152, 194, 169, 172, 179, 180, 153, 175]
y=[51, 44, 71, 65, 69, 88, 43, 66, 124.5, 76, 100, 50, 74, 55, 70, 65, 53, 60, 47, 100, 95, 58, 47, 70, 57, 67, 70, 82, 52, 85]

plt.scatter(x, y)
plt.show()
```

```
def cov(a, b):   #Função de covariância

    if len(a) != len(b):
        return

    a_mean = np.mean(a)
    b_mean = np.mean(b)

    sum = 0

    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))

    return sum/(len(a)-1)
```

```
covpeso_altura = cov(x, y)

coef_corrpeso_altura = covpeso_altura/(stcs.stdev(x)*stcs.stdev(y))

print("A covariância do peso e da altura é dada por %f" %covpeso_altura)
print("O coeficiente de correlação do peso e da altura é dada por %f" %coef_corrpeso_altura)
```
