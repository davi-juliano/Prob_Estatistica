# Exercício 3 -

3) Um fabricante de pentes de memória RAM para computadores, que produz em grandes quantidades, deseja estimar a fração 𝑝 de unidades defeituosas elaboradas por sua indústria. Para isso, ele selecionou uma amostra aleatória de 200 unidades e verificou que, entre elas, 5 eram defeituosas. Determine o intervalo de confiança otimista de 95% para 𝑝.

gama = 0.95/2 = 0.475

Z = 1.96

pchapeu = 5/200 = 0.025

`IC = [0.025-1.96*sqrt(0.025*(1-0.025)/200); 0.025+1.96*sqrt(0.025*(1-0.025)/200)] = [0.0033; 0.046] = [0.3%, 4,7%] `

```
print(0.95/2)

print(0.025 + 1.96*np.sqrt(0.025*(1-0.025)/200))
```
